```python
import math
import numpy as np
def bigprime(p):
    primel = []
    for i in range(2,p+1):
        prime = 1
        k = math.ceil(np.sqrt(i))
        while k not in primel and k>0:
            k -= 1
            if k == 1:             ##line specifically for 2 
                primel.append(i)   ##line specifically for 2
                k = i
        shortprimel = primel[0:primel.index(k)+1]
        for j in shortprimel:
            if i%j==0:
                prime = False
        if prime == True:
            if i not in primel:    ##line specifically for 2
                primel.append(i)
    return primel

def prime3(h):
    primelist = []
    for x in range(2,h+1):
        prime = 1
        for y in primelist:
            if x%y == 0:
                prime = 0
        if prime==1:
            primelist.append(x)
    return primelist

def prime4(h): # h must be even
    primelist = [2]
    for x in range(3,h+1,2):
        prime = 1
        for y in primelist:
            if x%y == 0:
                prime = 0
        if prime==1:
            primelist.append(x)
    return primelist

def bigprime2(p):
    primel = [2]
    for i in range(3,p+1,2):
        k = math.ceil(np.sqrt(i))
        while k not in primel:
            k -= 1
        if sum([i%j==0 for j in primel[0:primel.index(k)+1]])==0:
            primel.append(i)
    return primel

def newprime3(h):
    primelist=[2] #our end result that we also use throughout to generate the list
    endvalue=0 #the ID of the highest value we need to test from the list (the element that is greater than the sqrt of n)
    for x in range(3,h+1,2):
        prime=1
        for index,element in enumerate(primelist[endvalue:-1]):
            #print(index, element)
            if (element>math.sqrt(x)):
                endvalue=element
                #print(x, endvalue)
                break
        for y in primelist[0:endvalue]:
            if(x%y==0):
                prime=0
                break
        if (prime==1):
            primelist.append(x)
    return(primelist)   
   

def newprime4(h):
    #h is the highest number to be evaluated
    primelist=[2] #our end result that we also use throughout to generate the list
    endprime=2 #the ID of the highest value we need to test from the list (the element that is greater than the sqrt of n)
    for x in range(3,h+1,2):
        prime=1
        for index,element in enumerate(primelist[primelist.index(endprime):-1]):
            #print(index, element)
            if (element>math.sqrt(x)):
                endprime=element
                #print(x, endvalue)
                break
        for y in primelist[0:primelist.index(endprime)]:
            if(x%y==0):
                prime=0
                break
        if (prime==1):
            primelist.append(x)
    return(primelist)

def bigprime3(N):
    primel = [2]
    endindex = 0
    for i in range(3,N+1,2):
        for j in primel[endindex:-1]:
            if j > np.sqrt(i):
                endindex = primel.index(j)
                break
        #if sum([i%j==0 for j in primel[0:endindex]])==0: #does the same as below in 2 lines but its slower
         #   primel.append(i)
        prime = True
        for k in primel[0:endindex]:
            if i%k == 0:
                prime = False
                break
        if prime:
            primel.append(i)
    return primel
```
