```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def primes(N):
    primel = [2]
    endindex = 0
    for i in range(3,N+1,2):
        for j in primel[endindex:-1]:
            if j > np.sqrt(i):
                endindex = primel.index(j)
                break
        prime = True
        for k in primel[0:endindex]:
            if i%k == 0:
                prime = False
                break
        if prime:
            primel.append(i)
    return primel
#print(primes(200))
Q = 250

def spiral(turns):
    xs = [0]
    ys = [0]
    length =  1
    up = 1 #up and right = 1, down and left = -1
    for i in range(int(turns/2)):
        for j in range(length):
            xs.append(xs[-1] + up)
            ys.append(ys[-1])
        for k in range(length):
            xs.append(xs[-1])
            ys.append(ys[-1] + up)
        length += 1
        up *= -1            
    return xs,ys
xs,ys = spiral(Q)
plt.figure(figsize=(8,8))
plt.plot(xs,ys)
for p in primes(len(xs)):
    plt.scatter(xs[p-1],ys[p-1],color="green")
#plt.xlim(-8,8)
#plt.ylim(-8,8)
```
