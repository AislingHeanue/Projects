```python
def coprime(m,n):  #where m >= n
    while n!=0:
        m,n = n,m%n
    return m
def sumcoprimes(m):
    total = 0
    for n in range(m):
        if coprime(m,n) == 1:
            #print(n)
            total += n
    return total
sumcoprimes(987820)
```
